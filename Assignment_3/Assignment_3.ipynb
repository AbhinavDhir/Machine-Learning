```python
from sklearn.impute import SimpleImputer
```

```python
df.head()
```

```python
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='mean') 
df['AvgMonthSpend'] = mean_imputer.fit_transform(df[['AvgMonthSpend']].values)
```

```python
df.isnull().sum()
```

```python
mode_imputer = SimpleImputer(strategy='most_frequent')
df['BikeBuyer'] = mode_imputer.fit_transform(df[['BikeBuyer']])
```

```python
df.isnull().sum()
```

```python
#b
bins = np.linspace(min(df['YearlyIncome']), max(df['YearlyIncome']), 4) 
group_names = ['lower','lower middle','upper middle','upper'] 
```

```python
df['YearlyIncome'] = pd.cut(df['YearlyIncome'], bins, labels=group_names)
```
